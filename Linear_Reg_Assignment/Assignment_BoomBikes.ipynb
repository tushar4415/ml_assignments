### Data Preparation

```python
#Drop unwanted columns
inp0.drop(['instant','dteday','holiday','atemp','casual', 'registered'], axis = 1, inplace = True)
inp0.head()
```

```python
# Map Season String based on the values
# Defining the map function
def season_map(x):
    return x.map({1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'})

# Applying the function to the housing list
inp0[['season']] = inp0[['season']].apply(season_map)
```

```python
# Map weathersit String based on the values
# Defining the map function
def weathersit_map(x):
    return x.map({1: 'Clear', 2: 'Mist', 3: 'Light Snow', 4: 'Heavy Rain'})

# Applying the function to the housing list
inp0[['weathersit']] = inp0[['weathersit']].apply(weathersit_map)
```

```python
# Map Weekday String based on the values
# Defining the map function
def weekday_map(x):
    return x.map({0:'Sun',1:'Mon',2:'Tue',3:'Wed',4:'Thu',5:'Fri',6:'Sat'})

# Applying the function to the housing list
inp0[['weekday']] = inp0[['weekday']].apply(weekday_map)
```

```python
# Map Month String based on the values
# Defining the map function
def months_map(x):
    return x.map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})

# Applying the function to the housing list
inp0[['mnth']] = inp0[['mnth']].apply(months_map)
```

```python
inp0.head(5)
```

```python
# Create Dummy Variables and drop the first column using 'drop_first = True'
seasons = pd.get_dummies(inp0['season'], drop_first = True)
months = pd.get_dummies(inp0['mnth'], drop_first = True)
weekdays = pd.get_dummies(inp0['weekday'], drop_first = True)
weathersits = pd.get_dummies(inp0['weathersit'], drop_first = True)

# Add the results to the original housing dataframe
inp0 = pd.concat([inp0, seasons, months, weekdays, weathersits], axis = 1)
```

```python
inp0.head(5)
```

```python
#Drop columns as dummies are created.
inp0.drop(['season','mnth','weekday','weathersit'], axis = 1, inplace = True)
inp0.head()
```

## Splitting into Training and Testing datasets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(inp0, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescaling the Features using MinMax method

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to the required columns
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
X_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model
lm = sm.OLS(y_train,X_train_rfe).fit()

#Let's see the summary of our linear model
print(lm.summary())
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Drop const  
X_train_rfe.drop(['Jan'], axis = 1, inplace = True)
X_train_rfe.head()
```

```python
# Running the linear model
lm_2 = sm.OLS(y_train,X_train_rfe).fit()

#Let's see the summary of our linear model
print(lm_2.summary())
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Residual Analysis of the train data

```python
y_train_cnt = lm_2.predict(X_train_rfe)
```

```python
# Importing the required libraries for plots.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)   
```

### Making Predictions

```python
#Applying the scaling on the test sets
num_vars = ['temp', 'hum', 'windspeed', 'cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
#Dividing into X_test and y_test
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Now let's use our model to make predictions.
# Adding a constant variable 
X_test = sm.add_constant(X_test)

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_rfe.columns]
```

```python
# Making predictions
y_pred = lm_2.predict(X_test_new)
```

### Model Evaluation

```python
#Evaluate R-square for test
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
